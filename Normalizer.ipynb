```python
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import StandardScaler
import os
import numpy as np
import pandas as pd
import csv
```

```python
#spw = samples per window
def normalize(file, out, spw):
    #get data from file
    print("Reading data from: ", file)
    data = pd.read_csv(file,sep=',',header=None,dtype=np.float32)

    # get features columns
    x_data = data.iloc[:, 0:-1]
    # get label column
    y_data = data.iloc[:,[-1]]
    
    print("x_data shape:")
    print(x_data.shape)
    
    # ge features number
    n_features=int(x_data.shape[1]/spw)
    #init data
    data = np.zeros((x_data.shape[0],x_data.shape[1]))

    for i in range(n_features):
        scaler = MinMaxScaler()
        #scaler = StandardScaler()
        feat = scaler.fit_transform(x_data.iloc[:,i::n_features])
        for l in range(feat.shape[1]):
            data[:,(l*n_features)+i] = feat[:,l]
            data[:,(l*n_features)+i] = feat[:,l]
    data = np.hstack((data,y_data))
    print("Writing normalized data to: ", out)
    np.savetxt(out, data, fmt='%.6f', delimiter=",")

def normalize_subjects(path_in, prefix, path_out, suffix, n_subjects):
    #Normalization 
    for j in range(1,n_subjects+1):
        normalize((path_in + prefix +str(j)+'.csv'),path_out + prefix + str(j) + suffix + '.csv',1)
```

```python
normalize_subjects("../All_csv_subjects/Biphase_subjects/clean/", "Biphase_Subject", "../All_csv_subjects/Biphase_subjects/min-max/", "_norm",26)
```
